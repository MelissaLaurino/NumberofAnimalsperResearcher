R Script to search for phrases or text in string cells within multiple different columns and save the results as a dataframe and .csv file.

```r
#Load the library
library(dplyr)

#Read the master data file into a dataframe
df <- read.csv("2014_2023_Master_Laurino.csv")

#Select rows where "Carlee" is present in any of the specified columns
filtered_df <- df %>%
  filter(grepl("Carlee", Naturalist, ignore.case = TRUE) |
         grepl("Carlee", DataCollector, ignore.case = TRUE) |
         grepl("Carlee", Photographer, ignore.case = TRUE) |
         grepl("Carlee", DataEntered, ignore.case = TRUE))

#View the filtered dataframe
head(filtered_df$Date)
```

```r
#Save the filtered dataframe as a new .csv file.
write.csv(filtered_df, file = "CarleeX.csv", row.names = FALSE)
```

```r
#Read in the new .csv file
filtered_df <- read.csv("CarleeX.csv")
```

```r
#Subset Carlee's data to only include dolphins from the Species column
subset_df <- filtered_df %>%
  filter(Species == "T.t.")
```

```r
#Calculate the average of the Min and Max columns and save the results as a new column AverageGS
subset_df$AverageGS <- rowMeans(subset_df[, c("Min", "Max")], na.rm = TRUE)
```

```r
#Print the new column created within the subset dataframe:
subset_df$AverageGS
```

```r
#Calculate the sum of all the rows of the AverageGS column
SUM_AverageGS <- sum(subset_df$AverageGS, na.rm = TRUE)

#Print the sum
print(SUM_AverageGS)
```
